```python
# now we have a list agianst each movie the pandas will thiink of it ass a list and messes up genres thinking of each combination as a
# specific genre. so we will explode the dataset
df_exploded = df.explode('genres')
```

```python
#Lets see the hgeest rated Genre by average
df_exploded.groupby("genres")["vote_average"].mean().sort_values(ascending=False)
```

```python
# lets see average revenue per genre
df_exploded.groupby("genres")["revenue"].mean().sort_values(ascending=False)
```

```python
#lets make a genre by year table to ssee which genre was most common in which year
genre_trends=df_exploded.groupby(["release_year","genres"]).size().unstack(fill_value=0)
```

```python
genre_trends
```

```python
genre_summary=df_exploded.groupby("genres").agg({
    'revenue':'mean',
    'vote_average':'mean'
}).sort_values('revenue',ascending=False)
```

```python
import matplotlib.pyplot as plt
```

```python
import matplotlib.pyplot as plt

# Scale revenue into millions
genre_summary['revenue_millions'] = genre_summary['revenue'] / 1_000_000

plt.figure(figsize=(12, 7))
plt.scatter(
    genre_summary['vote_average'], 
    genre_summary['revenue_millions']
)

# Add labels for each genre
for genre in genre_summary.index:
    plt.text(
        genre_summary.loc[genre, 'vote_average'],
        genre_summary.loc[genre, 'revenue_millions'],
        genre,
        fontsize=8
    )

plt.xlabel("Average Rating")
plt.ylabel("Average Revenue (Millions)")
plt.title("Revenue vs Rating by Genre")
plt.show()
```
